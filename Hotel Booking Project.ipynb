```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
segments = ['Online', 'Offline']
p_c_values = [p_online, p_offline]
p_not_c_values = [1 - p_online, 1 - p_offline]

# Setting up the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Defining the position of the bars
x = np.arange(len(segments))
width = 0.35

# Plotting the bars
bars1 = ax.bar(x - width/2, p_c_values, width, label='P-C', color='blue')
bars2 = ax.bar(x + width/2, p_not_c_values, width, label='P-not-C', color='orange')

# Adding labels and title
ax.set_xlabel('Market Segment Type')
ax.set_ylabel('Probability')
ax.set_title('Probability Distribution for P-C and P-not-C')
ax.set_xticks(x)
ax.set_xticklabels(segments)
ax.legend()

# Adding value labels on the bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords='offset points',
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)

# Display the plot
plt.show()
```

```python
# Calculate probability of 'Offline'
p_offline = df[df['market segment type'] == 'Offline'].shape[0] / df.shape[0]

# Calculate probability of not 'Offline'
p_not_offline = 1 - p_offline

print(f"P-Offline: {p_offline * 100}%")
print(f"P-not-Offline: {p_not_offline * 100}%")

```

```python
# Calculate probability of 'Online'
p_online = df[df['market segment type'] == 'Online'].shape[0] / df.shape[0]

# Calculate probability of not 'Online'
p_not_online = 1 - p_online

print(f"P-Online: {p_online * 100}%")
print(f"P-not-Online: {p_not_online * 100}%")
```
